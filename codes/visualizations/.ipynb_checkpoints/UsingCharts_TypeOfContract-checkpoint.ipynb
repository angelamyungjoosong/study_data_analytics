```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font',family='Malgun Gothic')
import seaborn as sns
```

```python
import pandas as pd
```

```python
df_TOCC = pd.read_csv('../../../datasets/TypeOfContractChannel.csv')
df_TOCC[:2]
```

#### 다변수

#### X(설명변수) / Y(목표변수) : 범주형 / 연속형 

```python
df_TOCC.columns
```

```python
df_TOCC['amount'].describe()
## AMOUNT는 목표변수. 연속형. 
```

```python
# sns.barplot(data=df_TOCC, x='channel', y='amount')
#기본평균 (채널별 평균) 
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=sum)
#기본총합 / np.sum과 비슷 
plt.show()
#유의범위 거의 사용안함 
```

```python
#요약 툴 - 데이터의 특정 부분 요약 - group by - 어떤값을 기준으로 - 디폴트는 평균이 나와 - 집합함수인 aggfunc 
df_TOCC_sum = pd.pivot_table(data=df_TOCC, index='channel', values='amount', aggfunc=sum)
# 데이터 추출 마지막 7개, index만, list 형식으로 
df_TOCC_bottom_list = df_TOCC_sum.sort_values(by='amount', ascending=False).iloc[-7:].index.to_list()
#이걸 시각화 
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=sum, order=df_TOCC_bottom_list) #하위 7개 순위 표현 
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=sum, order=df_TOCC_bottom_list, hue='type_of_contract')
# 계약 종류 포함/ amount가 맴버쉽, 렌탈 중 어떻게 이루어져있는지 
```

```python
df_TOCC_top_list = df_TOCC_sum.sort_values(by='amount', ascending=False).iloc[:3].index.to_list()
df_TOCC_top_list
```

```python
sns.boxplot(data=df_TOCC, x='channel', y='amount', order=df_TOCC_top_list)
# 중앙값을 기준으로 Q1,Q3 내의 QCR 50퍼센트 볼수있다. 1.5 이상 이상치 볼 수 있다. 
# 중앙값이 안보이는 경우에는 박스의 경계에 있는 경우 (Q1이나 Q3과 중앙값이 겹쳐있는 경우) 
```

#### X(설명변수) / Y(목표변수) : 연속형 / 연속형 
둘다 숫자로 되어있는 경우 -> scatterplot사용 
- 상관성: 사회과학/마케팅(인간 개입됨) 높지 않음. 제조공정/실험실/자연과학 높음. 

```python
df_TOCC.describe()[:2]
```

```python
sns.scatterplot(data=df_TOCC, x='Mileage', y='age', hue='type_of_contract', size='Term')
# 나이에 따른 마일리지 
# 분석: 마일리지가 많을 수록 나이가 많다./ 25~35세 사이 고객 타겟팅 

# 특정 나이대를 타겟팅하고 싶다고 생각되면 이러한 패턴이 연령대별로 총합과도 맞느냐 확인 / 연령대를 범주형으로 만들어서 apply 활용 
# 렌탈 재계약율이 낮다는걸 알 수 있다 / 새계약이 많다 / 만족율이 낮다는 뜻 
# 재계약을 한 사람들은 기간을 짧게 / 신규고객을 위한 마켓팅 비용은 많이 들지만 재계약 즉 서비스측면은 떨어진다고 알 수 있다
# 개선점을 찾아야 함. 

# 연속형/ 연속형일때는 추세선 (변화상태를 선으로 그리는 것) 
# 나이에 따라 마일리지가 증가하는 걸 보여주는 선 필요(해당 나이의 점 분포의 중간에 있는 선을 찾아. 중앙의 선-머신러닝)
```

```python
sns.lmplot(data=df_TOCC, x='Mileage', y='age', hue='product', scatter=False)
# lm은 linear model / 개별적 값들의 중앙을 지나는 선 
# 단면으로 잘랐을때의 kde(확률 분포)를 그려서 각각의 가장 높은 값들끼리 연결해서 직선이 만들어진다.
# 각 분포를 직각으로 한 좌표의 분포도의 평균을 찾아 점으로 연결한 것 
# 제품의 차별없이 마일리지가 측정되구나 확인 가능 
```

```python
sns.jointplot(data=df_TOCC, x='Mileage', y='age', hue='product')
#단일변수가 같이 나옴 
```

```python
df_TOCC.corr()
# correlation 두가지 변수의 상관관계, 관계를 숫자로 정의 
# 이건 연속형 변수에만 적용 
# 비례 및 반비례를 통계적 수치로 나타낸 것 / 0인 경우 직선의 관계, -1에서 1 사이의 값 
# 0 으로 갈 수록 관계성이 없다. 
# 절대값 0.5 이상 이면관계성 있다/ 그 이하면 관계성 거의 없다고 봄. 
# 대각 행렬: 대각선을 기준으로 양쪽 값이 동일 대칭, 2차원 행렬 
# mileage와 age의 관계성이 가장 높다 
# 반비례로는 amount와 id가 그나마 가장 높지만, 거의 없다고 봄 
# correlation 검증 통해 관계성 여부를 어느정도 판단 (상대를 설득하는 도구 정도로만 사용되지만, 확실히 판단은 불가)
```
