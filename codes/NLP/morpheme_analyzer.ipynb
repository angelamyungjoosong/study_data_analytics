### Okt : 한글 형태분석기 중 하나 

```python
from konlpy.tag import Okt
```

```python
okt = Okt() #형태소 분석기 선택
```

```python
sentence_list = ['아버지가방에들어가신다'
                    ,'나는 밥을 먹는다.'
                    ,'하늘을 나는 자동차'
                    , '아이폰 기다리다 지쳐 애플 공홈에서 연락폰질러버렸다 6+ 128기가실버ㅋ']
```

```python
morphs_list = okt.morphs(sentence_list[1]) #토큰화
morphs_list 
#내부에 변수로 가지고 있음 -> 형태소 분석 후에 값으로 들어감 
```

```python
#품사 달기 
okt.pos(sentence_list[1])
#리스트에 듀플 형식으로 각 단어의 품사 
```

```python
okt.nouns(sentence_list[1])
#명사만 담기 
```

```python
okt.phrases(sentence_list[3])
#의미상으로 묶일 수 있는 여러방식을 보여줌 
# ex)애플 공홈 -> 애플 / 공홈 같이 분리될 수 있는 단어들 
```

```python
#결합명사 단어별 추출 
okt.phrases('애플공홈')
```

```python
from mecab import MeCab
```

```python
#인스턴스화
mecab = MeCab() 
```

```python
#토큰화
mecab.morphs(sentence_list[0])
```

```python
#품사 확인 
mecab.pos(sentence_list[0])
```

```python
mecab.parse('애플공홈')
```

```python
for sentence in sentence_list:
    morphs_list = mecab.morphs(sentence)
    print(morphs_list) 
```
