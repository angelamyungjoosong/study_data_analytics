```python
from mecab import MeCab
mecab = MeCab()
mecab.pos(comment_list[0])
```

### 형태소 분석기 통한 품사 분류
- 형태소 분석을 통해 리스트로 나온것을 문장화

```python
#품사 확인 및 리스트에 담기
list_comment_list_pos = list()
for comment_pos in comment_list:
    comment_list_pos = mecab.pos(comment_pos)
    list_comment_list_pos.append(comment_list_pos)
list_comment_list_pos
```

```python
#형태소 확인
for comment_morphs in comment_list:
    morphs_list_morphs = mecab.morphs(comment_morphs)
    print(morphs_list_morphs)
```

```python
#불용어
chosen_pos = ['MAG', 'NNG', 'NP', 'XR', 'NNP']
```

```python
#불용어 제거

comment_list_pos_without_stop = list()
for comment_word_pos in list_comment_list_pos:
  comment_word_pos_without_stop = list()
  for (word, pos) in comment_word_pos:
        if pos in chosen_pos:
            comment_word_pos_without_stop.append(word)

  comment_list_pos_without_stop.append(comment_word_pos_without_stop)
comment_list_pos_without_stop
```

### 백터 토큰화(단어 숫자화)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
comment_chosen_pos = list()
for comment_sentence_pos_without_stop in comment_list_pos_without_stop:
  comment_chosen_pos.append(' '.join(comment_sentence_pos_without_stop))
comment_chosen_pos
```

```python
len(comment_chosen_pos)
```

```python
#교육
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(comment_chosen_pos)
```

```python
tfidfVectorizer.idf_
```

```python
# 설명변수
features = tfidfVectorizer.transform(comment_chosen_pos)
```

### 목표 변수 생성

```python
target = label_list
target
```

### 머신러닝 학습

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(features.toarray(), target)
```

### 머신러닝 통한 예측

```python
comment = ' 너무너무감동적이야역시픽사담'
```

```python
tokenized_comment_morphs = mecab.morphs(comment)
tokenized_comment_morphs
```

```python
tokenized_comment_pos = mecab.pos(comment)
tokenized_comment_pos
```

```python
#불용어
chosen_pos = ['MAG', 'NNG', 'NP', 'XR', 'NNP']
```

```python
comment_list_pos_without_stop = list()

for comment_word_pos in tokenized_comment_pos:
    comment_word_pos_without_stop = list()

    word, pos = comment_word_pos

    if pos in chosen_pos:
        comment_word_pos_without_stop.append(word)

    comment_list_pos_without_stop.extend(comment_word_pos_without_stop)

# 결과 확인
print(comment_list_pos_without_stop)
```

```python
vectorized_comment = tfidfVectorizer.transform([' '.join( comment_list_pos_without_stop)])
vectorized_comment.toarray()
```

```python
model.predict(vectorized_comment.toarray()) #0=부정 1=긍정
#결과는 긍정
```
