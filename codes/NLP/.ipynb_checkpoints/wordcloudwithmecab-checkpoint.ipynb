## 자연어 처리 위한 단어 생성 

```python
# 영어: 단어별 space만 빼고 관사만 빼면 됨 
# 한국어: 조사가 있어서 그걸 제외시켜줘야함, 우선 space로 구분 

# "여행 계획을 세우고 있는데 어디로 갈까요?" -> 6개 
# "맛있는 음식을 먹으러 갈까요?", -> 4개

# 10개이지만 동일한 단어는 겹칠 수 있다. (갈까요) 
# 통으로 만들어 space로 잘라서 그 list에서 동일한 것은 count하고 나머지는 1 
# 단어별로 count되는 방식 
```

```python
# 앞에 space를 넣어서 sentences라는 list를 string으로 join한다 -> 한 문장으로 만든다
text = ' '.join(sentences)
text 
```

```python
# split해서 단어로 구성된 리스트 만들기 
text_split = text.split()
text_split
```

```python
from collections import Counter
Counter(text_split)
# 단어의 리스트를 카운터에 넣기 
```

```python
# 영어가 아닌 언어를 넣을 때는 폰트 연결 필요 (폰트 경로) 

# 한글 폰트 경로 설정 (Windows 환경에 맞게 경로 수정)
font_path = 'C:/Windows/Fonts/malgun.ttf'
wordCloud = WordCloud(font_path=font_path, background_color='white')
wordCloud.generate_from_text(text)
```

```python
wordCloud.words_
# wordCloud 내의 변수(어떤 단어가 몇퍼센트인지 인스턴스화 되어있음) 
# Count와 달리 퍼센트로 들어감 
```

```python
import matplotlib.pyplot as plt 
plt.imshow(wordCloud)
#축 없애거나 바탕색 바꾸기
plt.axis('off')
# 이미지 보여주기
plt.show()
```

## 자연어 처리 위한 형태분석기 적용 

```python
from mecab import MeCab 
mecab = MeCab()
```

```python
mecab.nouns(sentences[0])#mecab.pos(sentences[0])
```

```python
mecab.nouns(sentences[0]), mecab.nouns(sentences[1]),mecab.nouns(sentences[0])+ mecab.nouns(sentences[1])
# 두개의 리스트 붙여서 1차원으로 하나의 리스트 
# append 사용하면 2차원 리스트 
```

```python
# 형태분석기 적용해 명사만 추출 
morphs_list = list()
for sentence in sentences : 
    morphs = mecab.nouns(sentence)
    morphs_list = morphs_list + morphs 
morphs_list
# 대개 한글자는 제외시켜야함을 볼 수 있다
```

```python
from collections import Counter
nouns_count = Counter(morphs_list)
# 단어의 리스트를 카운터에 넣기 
```

```python
# 한글 폰트 경로 설정 (Windows 환경에 맞게 경로 수정)
font_path = 'C:/Windows/Fonts/malgun.ttf'
wordCloud = WordCloud(font_path=font_path, background_color='white')
wordCloud.generate_from_frequencies(nouns_count)
```

```python
wordCloud.words_
```

```python
import matplotlib.pyplot as plt 
plt.imshow(wordCloud)
plt.axis('off')
plt.show()
```
