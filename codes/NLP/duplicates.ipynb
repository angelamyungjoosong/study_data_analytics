```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
```

```python
df
```

```python
df.drop_duplicates() # 전체 컬럼 기준 중복값 드랍
```

```python
df.drop_duplicates(subset=['brand']) #brand를 기준으로 중복값 드랍 # 디폴트로 첫번째것이 남도록 된다
```

```python
df.drop_duplicates(subset=['brand'], keep='last')
#중복 중 마지막 것을 남길 때 사용
```

```python
df.drop_duplicates(subset=['brand', 'style'], keep='last')
```
